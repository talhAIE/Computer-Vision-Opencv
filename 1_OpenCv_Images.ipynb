#Function to Downlaod some images and unzip

```python
def download_and_unzip(url,save_path):
  #download
  urlretrieve(url,save_path)

  #unzip with exceptions
  try:
    with ZipFile(save_path) as z:
      z.extractall(os.path.split(save_path)[0]) #we are choosing only directory path here
    print("Unzipped Done\n")
  except Exception as e:
    print('Invalid File',e)
```

## Splitting and Merging Color Channels


* **`cv2.split()`** Divides a multi-channel array into several single-channel arrays.

* **`cv2.merge()`** Merges several arrays to make a single multi-channel array. All the input matrices must have the same size.

```python
#split imgage into bgr and display
b,g,r=cv2.split(img_bgr)
```

```python
#display images
plt.figure(figsize=[20,5])
plt.subplot(1,4,1);plt.imshow(b,cmap='gray');plt.title('Blue iamge')
plt.subplot(1,4,2);plt.imshow(g,cmap='gray');plt.title('green iamge')
plt.subplot(1,4,3);plt.imshow(b,cmap='gray');plt.title('red iamge')

#Merge individual channels into 1 BGR image
img_merge=cv2.merge((b,g,r))
plt.subplot(1,4,4);plt.imshow(img_merge[:,:,::-1]);plt.title('Merged Images and convert to RGB')
```

#Modifying Individual channle

```python
img=cv2.imread('/content/New_Zealand_Lake.jpg',cv2.COLOR_BGR2HSV)
h,s,v=cv2.split(img)
h_n=h+10
plt.figure(figsize=[20,5])
plt.subplot(1,4,1);plt.imshow(h_n,cmap='gray');plt.title('H channel')
plt.subplot(1,4,2);plt.imshow(s,cmap='gray');plt.title('S channel')
plt.subplot(1,4,3);plt.imshow(v,cmap='gray');plt.title('V channel')
img_nz_new=cv2.merge((h_n,s,v))[:,:,::-1]
plt.subplot(1,4,4);plt.imshow(img_nz_new);plt.title('original')
```

```python
cv2.imwrite('newModify.png',img_nz_new)
```
